```python
import requests, operator, glob2
from pandas import DataFrame
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime

def crawlingData():    
    l=[]#리스트    
    for page in range(50):       
        try:            
            r = requests.get("http://www.aladin.co.kr/shop/wbrowse.aspx?BrowseTarget=List&ViewRowsCount=50&ViewType=Detail&PublishMonth=0&SortOrder=4&page="
                             +str(page+1)+"&Stockstatus=1&PublishDay=84&CID=76730")            
            c = r.content
            soup = BeautifulSoup(c, "html.parser")
            all = soup.find("form", {"id": "Myform"})
            all2 = all.find_all("div",{"class":"ss_book_box"})       
            #전체를 찾아 all에 대입 후
            #all2에 배열 형태로 저장/하나씩            
            for item in all2:
                d={}                
                tmpitem = item.find("table",{"width":"100%"}).find("ul")
                tmpitem2=tmpitem.find_all("li")
                link=tmpitem2[4].find("a")["href"]
                index=link.index("=")
                index2=link.index("#")
                itemId=link[index+1:index2]
                l.append(itemId)
            #linkOpen(link, d)
            #l.append(d) #리스트에 사전 추가
            print("Loading...")
        except:
            print("예외가 발견되었습니다.")
    file_name='중등교재14.csv'
    df=pd.DataFrame(l)
    print(df)
    df.to_csv(file_name, encoding='utf-8-sig', index=False)
    print("파일이 정상적으로 생성되었습니다.")
crawlingData()
```

```python
import requests, operator, glob2
from pandas import DataFrame
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
import csv

def crawlingData():
    l=[]
    itemIdList=[]
    index=0
    getCsv(itemIdList)
    for itemId in itemIdList:
        try:
            r = requests.get("http://www.aladin.co.kr/ucl/shop/product/common/wa_Common_BookPleNewsFeed_Ajax.aspx?itemId="+itemId+"&viewRowCount=30")
            c = r.content
            soup = BeautifulSoup(c, "html.parser")
            all = soup.find_all("div",{"class":"p_feed_writer2"})
        #리뷰하나하나 all 저장
            for item in all:
                d={}
                split_li=item.find_all("li")
                img=split_li[0].find("img")["src"]
                get_score(img, d)
                review=split_li[1].text.replace("&nbsp","").replace("\r\n"," ").replace("\t"," ").replace("\n"," ")
                print(review)
                d['1.index']=index
                d['2.review']=review
                index+=1
                l.append(d)
            #linkOpen(link, d)
            #l.append(d) #리스트에 사전 추가
            print("Loading...")
        except:
            print("예외가 발견되었습니다.")
    df=pd.DataFrame(l)
    print(df)
    file_name='중등교재_14.csv'
    df.to_csv(file_name, encoding='utf-8-sig', index=False)
    #print("파일이 정상적으로 생성되었습니다.")
def get_score(img, d):
    if 'icon_star0' in img:
        d['3.score']=0
    elif 'icon_star1' in img:
        d['3.score']=1
    elif 'icon_star2' in img:
        d['3.score']=2
    elif 'icon_star3' in img:
        d['3.score']=3
    elif 'icon_star4' in img:
        d['3.score']=4
    elif 'icon_star5' in img:
        d['3.score']=5
def getCsv(itemIdList):
    f=open('중등교재14.csv','r',encoding='utf-8')
    #csv 파일오픈    
    csvReader=csv.reader(f)    
    for i in csvReader:
        itemIdList.append(i[0])
        #한 행씩 리스트에 저장 
    f.close()
crawlingData()
```
